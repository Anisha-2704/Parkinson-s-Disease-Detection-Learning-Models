```python
from pathlib import Path

BASE = Path(r"C:\Users\KIIT\Desktop\Parkinson_EEG\data")
DATA_RAW = BASE / "raw"

files = sorted(DATA_RAW.rglob("*.bdf"))  # <-- recursive listing

print("Total EEG files found:", len(files))
for f in files:
    print(" -", f.relative_to(DATA_RAW))
```

```python
hc_keywords = ["hc", "control"]
pd_keywords = ["pd", "patient"]

hc_files = [f for f in files if any(k in f.name.lower() for k in hc_keywords)]
pd_files = [f for f in files if any(k in f.name.lower() for k in pd_keywords)]

print("HC subjects :", len(hc_files))
print("PD subjects :", len(pd_files))

print("\nSample HC:")
for f in hc_files[:5]: print(" -", f.name)

print("\nSample PD:")
for f in pd_files[:5]: print(" -", f.name)
```
