```python
import mne
import numpy as np
import pandas as pd
from pathlib import Path

BASE = Path(r"C:\Users\KIIT\Desktop\Parkinson_EEG")
RAW = BASE / "data" / "raw"
OUT = BASE / "results1" / "SanDiego"
OUT.mkdir(parents=True, exist_ok=True)

SEG_LEN = 2       # seconds
SF = 256          # assume sampling rate (safe default)

def hjorth_params(sig):
    diff1 = np.diff(sig)
    diff2 = np.diff(diff1)
    var0 = np.var(sig)
    var1 = np.var(diff1)
    var2 = np.var(diff2)
    activity = var0
    mobility = np.sqrt(var1 / var0) if var0>0 else 0
    complexity = np.sqrt(var2 / var1) / mobility if var1>0 and mobility>0 else 0
    return activity, mobility, complexity

X, y, subjects = [], [], []

files = sorted(RAW.rglob("*.bdf"))
print("Total files:", len(files))

for f in files:
    print("Processing:", f.name)

    raw = mne.io.read_raw_bdf(f, preload=True, verbose=False)

    # ---------- ICA ONLY (no bandpass) ----------
    ica = mne.preprocessing.ICA(n_components=20, random_state=42)
    ica.fit(raw)
    raw_ica = ica.apply(raw.copy())

    data = raw_ica.get_data()
    n_samples = data.shape[1]
    step = SEG_LEN * SF

    n_segments = n_samples // step

    for i in range(n_segments):
        seg = data[:, i*step:(i+1)*step]

        feats = []
        for ch in seg:
            act, mob, comp = hjorth_params(ch)
            feats += [
                np.mean(ch), np.std(ch), act, mob, comp
            ]

        X.append(feats)

        label = 1 if "pd" in f.name.lower() else 0
        y.append(label)
        subjects.append(f.stem)

X = np.array(X, dtype=float)
y = np.array(y, dtype=int)
subjects = np.array(subjects)

print("Final shape:", X.shape)

np.savez(OUT/"features_caseA_ICA.npz",
         X=X, y=y, subjects=subjects, fs=SF)

print("\nSaved:", OUT/"features_caseA_ICA.npz")
```

```python
import mne
import numpy as np
from pathlib import Path

BASE = Path(r"C:\Users\KIIT\Desktop\Parkinson_EEG")
RAW = BASE / "data" / "raw" / "san diego"
OUT = BASE / "results1" / "SanDiego"
OUT.mkdir(parents=True, exist_ok=True)

SEG_LEN = 2   # seconds

def hjorth_params(sig):
    diff1 = np.diff(sig)
    diff2 = np.diff(diff1)
    var0 = np.var(sig)
    var1 = np.var(diff1)
    var2 = np.var(diff2)
    activity = var0
    mobility = np.sqrt(var1/var0) if var0>0 else 0
    complexity = (np.sqrt(var2/var1)/mobility
                  if var1>0 and mobility>0 else 0)
    return activity, mobility, complexity

X, y, subjects = [], [], []

files = sorted(Path(RAW).rglob("*.bdf"))
print("Total files:", len(files))

for f in files:
    print("Processing:", f.name)

    raw = mne.io.read_raw_bdf(f, preload=True, verbose=False)
    sf = raw.info["sfreq"]
    step = int(SEG_LEN * sf)

    # -------- ICA ONLY (no band-pass used for experiment) --------
    raw_tmp = raw.copy().filter(1., 40., fir_design="firwin",
                                verbose=False)  # for stable ICA fit only
    ica = mne.preprocessing.ICA(n_components=20, random_state=42)
    ica.fit(raw_tmp)
    raw_ica = ica.apply(raw.copy())

    data = raw_ica.get_data()
    n_samples = data.shape[1]
    n_segments = n_samples // step

    for i in range(n_segments):
        seg = data[:, i*step:(i+1)*step]

        feats = []
        for ch in seg:
            act, mob, comp = hjorth_params(ch)
            feats.extend([np.mean(ch), np.std(ch), act, mob, comp])

        X.append(feats)
        y.append(1 if "pd" in f.name.lower() else 0)
        subjects.append(f.stem)

X = np.array(X, float)
y = np.array(y, int)
subjects = np.array(subjects)

print("Final shape:", X.shape)

np.savez(OUT/"features_caseA_ICA.npz",
         X=X, y=y, subjects=subjects)

print("\nSaved:", OUT/"features_caseA_ICA.npz")
```
